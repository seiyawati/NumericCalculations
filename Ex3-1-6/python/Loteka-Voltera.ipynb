```python
import numpy as np
import matplotlib.pyplot as plt

# Lotka-Volterra Model Parameters
a = 1.0
b = 2.0
c = 1.0
d = 3.0

# Sample Time
h = 0.01

# Simulation Time
N = 3000
t = np.arange(N)*h

# Initial Values
x0 = 2.0
y0 = 1.0

# Dynamics of The Model
def f(x, y):
    kx = a*x - b*x*y
    ky = d*x*y - c*y
    return kx, ky

# State Transition using Runge-Kutta Method
def next(x, y):
    kx1, ky1 = f(x, y)
    kx2, ky2 = f(x + kx1*h/2, y + ky1*h/2)
    kx3, ky3 = f(x + kx2*h/2, y + ky2*h/2)
    kx4, ky4 = f(x + kx3*h, y + ky3*h)
    xnew = x + (kx1 + 2*kx2 + 2*kx3 + kx4)*h/6
    ynew = y + (ky1 + 2*ky2 + 2*ky3 + ky4)*h/6
    return xnew, ynew

# Simulation Loop
x = np.zeros(N)
y = np.zeros(N)
x[0] = x0
y[0] = y0
for k in range(N-1):
    x[k+1], y[k+1] = next(x[k], y[k])

# Plot Results
plt.subplot(1,2,1)
plt.plot(t,x, label='Prey')
plt.plot(t,y, label='Predator')
plt.grid()
plt.legend()
plt.xlabel('Time')

plt.show()
```
